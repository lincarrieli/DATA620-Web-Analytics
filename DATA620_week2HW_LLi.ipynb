```python
# import libraries
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# plot Krackhardt Kite Social Network
G = nx.krackhardt_kite_graph(create_using = None)
nx.draw_networkx(G)
plt.show()
```

```python
# get nodes 
list(G.nodes)
```

```python
# get edges
list(G.edges)
```

```python
# Relabel the nodes of the graph 
G = nx.krackhardt_kite_graph(create_using = None)
mapping = {0: "Andre", 1: "Beverley", 2: "Carol", 3: "Diane", 4: "Ed", 5: "Fernando", 6: "Garth", 7: "Heather", 8: "Ilk", 9: "Jane"}
H = nx.relabel_nodes(G, mapping)
nx.draw_networkx(H)
plt.show()
```
