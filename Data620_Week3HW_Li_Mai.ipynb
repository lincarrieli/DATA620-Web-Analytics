```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# load netork data
file = pd.read_csv('https://raw.githubusercontent.com/lincarrieli/DATA620-Web-Analytics/main/insecta-beetle-group-t1-period-1.txt', sep=" ", header=None)

# get edges columns
edges = pd.DataFrame(file[file.columns[0:2]])

# convert dataframe to edges list
edges_list = edges.values.tolist()

            
```

```python
# graph network
G = nx.to_networkx_graph(edges_list, create_using=nx.Graph())
print(nx.info(G))
nx.draw(G)
plt.show(G)
```

```python
# explore network metrices
# 1. diameter
diameter = nx.diameter(G, usebounds = False)
diameter
```

```python
# 2. density
density = nx.density(G)
density
```

```python
# 3. degree centrality
degree_centrality = nx.degree_centrality(G)
degree_centrality
```
